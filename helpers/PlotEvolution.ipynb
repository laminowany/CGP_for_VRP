```python
import pandas as pd
import re
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def parse_tensor(x):
    if isinstance(x, str):
        m = re.search(r"tensor\(([-+]?\d*\.?\d+)\)", x)
        if m:
            return float(m.group(1))
    return float(x)

df_baseline = pd.read_csv("/home/piotr/repos/magisterka/outputs/csv/vrp10_baseline.csv")
df_baseline["score"] = df_baseline["score"].apply(parse_tensor)
df_baseline["source"] = "baseline"

df_evol1 = pd.read_csv("/home/piotr/repos/magisterka/outputs/najlepszy_osobnik_pelnej_ewolucji_fullrun/epochs.csv")
df_evol1["score"] = df_evol1["score"].apply(parse_tensor)
df_evol1["source"] = "baseline"

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_baseline["epoch"],
    y=df_baseline["score"],
    mode="lines+markers",
    name=f"baseline"
))
fig.add_trace(go.Scatter(
    x=df_evol1["epoch"],
    y=df_evol1["score"],
    mode="lines+markers",
    name=f"evolution candidate #1"
))
fig.show(config={"scrollZoom": True})
fig.write_html("evolutionCandidate1.html" ,config={"scrollZoom": True})
```
